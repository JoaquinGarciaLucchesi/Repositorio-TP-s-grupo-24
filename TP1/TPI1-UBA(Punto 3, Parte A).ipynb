```python
df = pd.read_excel('data.xlsx')

df_arg = df[df['country'] == 'Argentina']
df_uy = df[df['country'] == 'Uruguay']

y1 = df_arg[df_arg['country'] == 'Argentina']['Producto Interno Bruto (PIB) en USD']
x1 = df_arg[df_arg['country'] == 'Argentina']['date']

y2 = df_arg[df_arg['country'] == 'Argentina']['Tasa de Crecimiento del PIB (%)']
x2 = df_arg[df_arg['country'] == 'Argentina']['date']

y3 = df_uy[df_uy['country'] == 'Uruguay']['Producto Interno Bruto (PIB) en USD']
x3 = df_uy[df_uy['country'] == 'Uruguay']['date']

y4 = df_uy[df_uy['country'] == 'Uruguay']['Tasa de Crecimiento del PIB (%)']
x4 = df_uy[df_uy['country'] == 'Uruguay']['date']
```

```python

plt.plot(x1, y1, label="PIB Arg", color='blue')
plt.plot(x3, y3, label="PBI Uru", color='orange')

# Modifico  labels
plt.xlabel("Fechas")
plt.ylabel("PBI")
plt.title("PBI en Dólares")

# Agrego  leyenda
plt.legend(loc='upper left')
plt.show() #esto es necesario para visualizar
```

```python
plt.plot(x2, y2, label="Tasa Arg", color='red')
plt.plot(x4, y4, label="Tasa Uru", color='green')

# Modifico  labels
plt.xlabel("Fechas")
plt.ylabel("Tasa en %")
plt.title("Tasa de Crecimiento del PBI %")

# Agrego  leyenda
plt.legend(loc='upper left')
plt.show() #esto es necesario para visualizar
```

```python
# Resolver acá estilo orientado-objetos 
# Tip: aprovechar este estilo de programar una figura para hacerlo más lindo 
fig, ax = plt.subplots()

ax.plot(x1, y1, label="PIB Arg", color='blue')
ax.plot(x2, y2, label="Tasa % PBI Arg", color='red')
ax.plot(x3, y3, label="PBI Uru", color='orange')
ax.plot(x4, y4, label="Tasa % PBI Uru", color='green')

ax.set_xlabel("Fechas")
ax.set_ylabel("PBI y Tasa Crec PBI %")
ax.set_title("PBI en dólares y Tasa de Crecimiento en %")

ax.legend()
fig.show()
```
